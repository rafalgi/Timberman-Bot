# Collecting screenshots of the game

```python
import uuid
import time
import os
import pyautogui
import cv2
import numpy as np
import mss

dimensionsLeft = {
        'top': 50,
        'left': 650,
        'width': 400,
        'height': 800
}

dimensionsRight = {
        'top': 50,
        'left': 1000,
        'width': 400,
        'height': 800
}
   
IMAGES_PATH = os.path.join('data', 'images')
labels = ['left_log', 'right_log', 'timberman']
number_imgs = 40


with mss.mss() as sct:

    for img_num in range(number_imgs):
        print(f'Collecting image number {img_num}')
        
        # Capture screenshot
        screen = sct.grab(dimensionsRight)
        
        # Convert screenshot to a numpy array
        screen_array = np.array(screen)
        
        # Convert colors from RGB to BGR (OpenCV uses BGR by default)
        corrected_colors = cv2.cvtColor(screen_array, cv2.COLOR_RGB2BGR)
        
        # Generate a unique filename
        imgname = os.path.join(IMAGES_PATH, f'{uuid.uuid1()}.jpg')
        
        # Save the image
        cv2.imwrite(imgname, corrected_colors)
        
        # Wait for 2 seconds
        time.sleep(2)
```

```python
# Part of the screen to capture
monitor = {
    "top": 40, 
    "left": 0, 
    "width": 1920, 
    "height": 1080
}

dimensionsLeft = {
        'top': 50,
        'left': 600,
        'width': 400,
        'height': 800
}

dimensionsRight = {
        'top': 50,
        'left': 1000,
        'width': 355,
        'height': 800
}
    

# Threshold
threshold = 0.71

# Flag to determine which side to monitor first (left side)
left = True

# Record the current time (for performance monitoring)

last_time = time.time()
keyboard.wait('s')

with mss.mss() as sct:
    


    while True:


        # Capture the left or right area of the screen based on the 'left' flag
        if left:
            img = np.asarray(sct.grab(dimensionsLeft))
            
        else:
            img = np.array(sct.grab(dimensionsRight))
            
        
          
        img = cv2.cvtColor(img, cv2.COLOR_RGB2BGR)
        
            
        
        results = model_v1(img)[0]

        
        for result in results.boxes.data.tolist():
            
            # Get the coordinates of the box
            x1, y1, x2, y2, score, class_id = result
            
        
            if score > threshold:
                if (class_id == 1  and  left == False) or (class_id == 0 and left == True ):
                    pass
            else:              
                if class_id == 0 :
                    
                    cv2.rectangle(img, (int(x1), int(y1)), (int(x2), int(y2)), (0, 255, 0), 4)
                    cv2.putText(img, results.names[int(class_id)].upper(), (int(x1), int(y1 - 10)),
                                cv2.FONT_HERSHEY_SIMPLEX, 1.3, (0, 255, 0), 3, cv2.LINE_AA)
                               
                        
                    left = not left
                            
                elif class_id == 1 : 
                    
                    cv2.rectangle(img, (int(x1), int(y1)), (int(x2), int(y2)), (0, 255, 0), 4)
                    cv2.putText(img, results.names[int(class_id)].upper(), (int(x1), int(y1 - 10)),
                                cv2.FONT_HERSHEY_SIMPLEX, 1.3, (0, 255, 0), 3, cv2.LINE_AA)
                    left =True     
        
        if left:
            pyautogui.press('left')
        else:
            pyautogui.press('right')
        


        
        img = cv2.cvtColor(img , cv2.COLOR_BGR2RGB)
        cv2.imshow("OpenCV/Numpy normal", img)
    

        
        if cv2.waitKey(25) & 0xFF == ord("q"):
            cv2.destroyAllWindows()
            break

        time.sleep(0.01)
```
